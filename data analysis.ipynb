TOP DIRECTORS

```python
top_rating = df['rating'].value_counts().idxmax()
print("Most common rating:", top_rating)

filtered = df[df['rating'] == top_rating].copy()

filtered['director'] = filtered['director'].str.split(',')
filtered = filtered.explode('director')
filtered['director'] = filtered['director'].str.strip()

top_directors = filtered['director'].value_counts().head(10)
top_directors = top_directors.drop(index="unknown", errors="ignore")

plt.figure(figsize=(10,6))
top_directors.plot(
    kind="pie",
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Purples(range(50, 50 + len(top_directors)*20, 20))
)

plt.title(f"Top Directors for Rating: {top_rating}")
plt.tight_layout()
plt.show()
```

INSIGHT

Jan Suter has the highest number of TV-MA titles. Raul Campos is the second highest. There is a clear drop after the top two directors. Most other directors have fewer and similar number of titles.

TOP 10 DIRECTORS WITH MOST MOVIES IN NETFLIX

```python
top_directors=df['director'].value_counts().head(10)
top_directors=top_directors.drop(index="unknown")
plt.figure(figsize=(10,6))
top_directors.plot(kind="bar")
plt.title("top 10 directors with the Most Movies on Netflix")
plt.xlabel('Directors')
plt.ylabel('Number of movies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

NUMBER OF MOVIES SHOT IN UNITED STATES IN NETFLIX

```python
us_movies=df[df['country']=='United States']
movies_per_year=us_movies['release_year'].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.plot(movies_per_year.index,movies_per_year.values,marker='o')
plt.title("Number of Movies Shot in United states(year-wise)")
plt.xlabel("Release_year")
plt.ylabel("Number of Movies")
plt.grid()
plt.tight_layout()
plt.show()
```

INSIGHTS

The number of movies was very low before 2000. Movies started to increse slowly after 2000. There is a big rise after 2010. The highest number of movies is around 2018-2020. After the peak, the number of movies drops suddenly.

RELEASE YEAR VS COUNTRY(TOP 5 COUNTRIES)

```python
top_countries=df['country'].value_counts().head(5).index
gf=df[df['country'].isin(top_countries)]
pivot=gf.pivot_table(index='release_year',columns='country',aggfunc='size',fill_value=0)
plt.figure(figsize=(12,6))
for country in pivot.columns:
    plt.plot(pivot.index,pivot[country], marker='o', label=country)


plt.title("Release Year vs Country (Top 5 Countries)")
plt.xlabel("Release Year")
plt.ylabel("Number of Movies")

plt.legend()
plt.grid()
plt.tight_layout()
```

INSIGHTS

All countries had very few movies before 2000. After 2000, movies start increasing in every country. The United States has the highest number of movies. Countries like India, United Kingdom and Canada also shows the growth after 2010. Most countries reach their peak around 2018-2020 then slightly decreses.

## 1. How has the number of movies vs TV shows added to Netflix changed overtime?

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year
data = df.groupby(['year_added', 'type']).size().unstack()
data.plot(figsize=(10,5))
plt.title('Movies vs TV Shows Added Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

OBSERVATIONS

- Movies are much higher in number compared to TV shows.

- After around 2015, both movies and TV shows increased rapidly.

- Netflix started adding more TV shows in recent years but movies still dominate . 

INSIGHT

Netflix is shifting slowly toward more series content. But movies are still the majority.

## 2. Which countries produce the most Netflix content?

```python
df['country'] = df['country'].str.split(',').str[0]
df['country'] = df['country'].str.strip().str.lower()
df = df.dropna(subset=['country'])
df = df[df['country'] != 'unknown']
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(10,6))
top_countries.plot(kind='bar')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

OBSERVATIONS

- United States is the highest country which produce more Netflix content.

- India is the second one.

- United Kingdom, Canada, Japan also significant.

INSIGHT

Netflix content is heavily dominated by the United States, but India is also a major contributor.

## 3. What is the most common genres on Netflix?

```python
genres = df['listed_in'].str.split(',', expand=True).stack()
top_genres = genres.value_counts().head(10)
top_genres.plot(kind='bar')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

OBSERVATIONS
